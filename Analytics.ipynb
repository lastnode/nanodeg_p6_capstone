## Lichess

```python
files2 = glob.glob('output_data/fact/lichess/moves7/*.parquet')
df_lichess = pd.concat([pd.read_parquet(fp) for fp in files2])
```

```python
df_lichess
```

```python
df_merged = pd.concat([df_chesscom, df_lichess])
```

```python
df_merged.reset_index(inplace=True)
```

```python
df_merged
```

## Compile Openings

```python
openings_query = """      select
                            opening as opening,
                            sum(case when winner = "white" then 1 else 0 end) as white_wins,
                            sum(case when winner = "black" then 1 else 0 end) as black_wins,
                            count(*) as game_count
                        from df_merged

                        group by opening
                        order by black_wins desc"""
```

```python
openings_df = ps.sqldf(openings_query, locals())
```

```python
openings_df
```

## Compile Players

```python
players_query = """ with white_players as (

                    select
                        white_username as username,
                        sum(case when winner = "white" then 1 else 0 end) as white_wins,
                        count(*) as white_games
                    from df_merged

                    group by white_username
                        ),
                        
                    black_players as (
                    
                    select
                        black_username as username,
                        sum(case when winner = "black" then 1 else 0 end) as black_wins,
                        count(*) as black_games
                    from df_merged

                    group by black_username
                    )
                    
                    select
                        white_players.username,
                        white_wins,
                        white_games,
                        cast(white_wins / white_games as float(10,10)) as white_win_perc,
                        black_wins,
                        black_games,
                        cast (black_wins / black_games as float(10)) as black_win_perc
                    from white_players
                    
                    inner join black_players on white_players.username = black_players.username
                    
                    order by white_wins desc
                      """
```

```python
players_df = ps.sqldf(players_query, locals())
```
